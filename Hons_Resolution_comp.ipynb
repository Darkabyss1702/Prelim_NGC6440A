```python
subints ={}

for i in range(1,306):
    filename = f"Profile_stokes_{i}.csv"
    subints[i] = np.loadtxt(filename, delimiter=",", skiprows=1)
```

```python
def Gauss(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
```

# DOUBLE MODE ANALYSIS
# Using double mode pulses (1,7,8,9,10,17)

```python
x_values = np.arange(1,257)
y_values = {}
popt_dict ={}
pcov_dict = {}
skewness ={}
kurt ={}
fontsize=8.5
sigma ={}

double_list = [1,7,8,9,10,17] 
color_double = 'royalblue'

for i in double_list:
    #print(i)
    y_values[i] = subints[i][:,0] #use y_values[i] for Stokes I of each subintegration i]
    popt_dict[i], pcov_dict[i] = curve_fit(Gauss, x_values, y_values[i], p0=[max(y_values[i]), np.mean(x_values), np.std(x_values)])
    skewness[i] = skew(y_values[i])
    kurt[i] = kurtosis(y_values[i])
```

```python
fig, ax = plt.subplots(3,2, figsize=(15,12))
ax[0][0].plot(x_values, y_values[1], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[0][0].plot(x_values, Gauss(x_values, *popt_dict[1]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[0][0].axhline(popt_dict[1][0], linestyle='--', alpha = 0.3)
ax[0][0].text(0, popt_dict[1][0], f"Peak: {popt_dict[1][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[0][0].legend()
ax[0][0].set_title("Double Mode fitted with Gaussian (Subint 1)")
ax[0][0].text(0.65, 0.65, f"Skewness: {skewness[1]:.2f}\nKurtosis: {kurt[1]:.2f}\nMean: {popt_dict[1][1]:.2f}\nStd.Dev: {popt_dict[1][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))


ax[0][1].plot(x_values, y_values[7], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[0][1].plot(x_values, Gauss(x_values, *popt_dict[7]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[0][1].axhline(popt_dict[7][0], linestyle='--', alpha = 0.3)
ax[0][1].text(0, popt_dict[7][0], f"Peak: {popt_dict[7][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[0][1].set_title("Double Mode fitted with Gaussian (Subint 7)")
ax[0][1].text(1.92, 0.9, f"Skewness: {skewness[7]:.2f}\nKurtosis: {kurt[7]:.2f}\nMean: {popt_dict[7][1]:.2f}\nStd.Dev: {popt_dict[7][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))


ax[1][0].plot(x_values, y_values[8], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[1][0].plot(x_values, Gauss(x_values, *popt_dict[8]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[1][0].axhline(popt_dict[8][0], linestyle='--', alpha = 0.3)
ax[1][0].text(0, popt_dict[8][0], f"Peak: {popt_dict[8][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[1][0].set_title("Double Mode fitted with Gaussian (Subint 8)")
ax[1][0].text(0.75, -0.3, f"Skewness: {skewness[8]:.2f}\nKurtosis: {kurt[8]:.2f}\nMean: {popt_dict[8][1]:.2f}\nStd.Dev: {popt_dict[8][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))


ax[1][1].plot(x_values, y_values[9], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[1][1].plot(x_values, Gauss(x_values, *popt_dict[9]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[1][1].axhline(popt_dict[9][0], linestyle='--', alpha = 0.3)
ax[1][1].text(0, popt_dict[9][0], f"Peak: {popt_dict[9][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[1][1].set_title("Double Mode fitted with Gaussian (Subint 9)")
ax[1][1].text(1.92, -0.3, f"Skewness: {skewness[9]:.2f}\nKurtosis: {kurt[9]:.2f}\nMean: {popt_dict[9][1]:.2f}\nStd.Dev: {popt_dict[9][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))


ax[2][0].plot(x_values, y_values[10], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[2][0].plot(x_values, Gauss(x_values, *popt_dict[10]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[2][0].axhline(popt_dict[10][0], linestyle='--', alpha = 0.3)
ax[2][0].text(0, popt_dict[10][0], f"Peak: {popt_dict[10][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[2][0].set_title("Double Mode fitted with Gaussian (Subint 10)")
ax[2][0].text(0.75, -1.50, f"Skewness: {skewness[10]:.2f}\nKurtosis: {kurt[10]:.2f}\nMean: {popt_dict[10][1]:.2f}\nStd.Dev: {popt_dict[10][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))


ax[2][1].plot(x_values, y_values[17], color = 'royalblue', alpha=0.8, label='psr pulse')
ax[2][1].plot(x_values, Gauss(x_values, *popt_dict[17]), label = 'fitted Gaussian', color = 'black', alpha=0.8)
ax[2][1].axhline(popt_dict[17][0], linestyle='--', alpha = 0.3)
ax[2][1].text(0, popt_dict[17][0], f"Peak: {popt_dict[17][0]:.3f}", fontsize=9, verticalalignment='bottom',
        horizontalalignment='left', fontstyle='italic', color='royalblue')
ax[2][1].set_title("Double Mode fitted with Gaussian (Subint 17)")
ax[2][1].text(1.92, -1.5, f"Skewness: {skewness[17]:.2f}\nKurtosis: {kurt[17]:.2f}\nMean: {popt_dict[17][1]:.2f}\nStd.Dev: {popt_dict[17][2]:.2f}", 
              transform=ax[0][0].transAxes, fontsize=fontsize, 
              verticalalignment='top', bbox=dict(facecolor='white'))
```
